```python
import random
import math
import matplotlib.pyplot as plt

def aleatorioRandom():
    return random.random()

def exponencial(valor_lambda):
    numero_aleatorio = aleatorioRandom()
    valor = - (math.log(1 - numero_aleatorio) / float(valor_lambda))
    return valor

def normal():
    numero_aleatorio = aleatorioRandom()
    raiz = math.sqrt(-2 * math.log1p(numero_aleatorio - 1))
    seno = math.sin(2 * math.pi * numero_aleatorio)
    valor = raiz * seno
    if(valor < 0):
        valor = 0
    return valor
```

```python
def llegada(tiempo_actual_simulado):
    global promedio_de_clientes
    global numero_de_clientes
    global numero_de_llegadas
    global llegadas
    global tiempo
    global valor_lambda
    global tiempo_maximo_de_servicio
    global tiempo_mayor
    global clientes, tiempos
    
    clientes['nodo 1'].append(numero_de_clientes['nodo 1'])
    tiempos['nodo 1'].append(tiempo)
    clientes['nodo 2'].append(numero_de_clientes['nodo 2'])
    tiempos['nodo 2'].append(tiempo)
    clientes['nodo 3'].append(numero_de_clientes['nodo 3'])
    tiempos['nodo 3'].append(tiempo)
    
    
    promedio_de_clientes['nodo 1'] += numero_de_clientes['nodo 1'] * (tiempo_actual_simulado - tiempo)
    numero_de_clientes['nodo 1'] += 1
    numero_de_llegadas['nodo 1'] += 1
    llegadas['nodo 1'].append(tiempo_actual_simulado)
    
    
    clientes['nodo 1'].append(numero_de_clientes['nodo 1'])
    tiempos['nodo 1'].append(tiempo)
    clientes['nodo 2'].append(numero_de_clientes['nodo 2'])
    tiempos['nodo 2'].append(tiempo)
    clientes['nodo 3'].append(numero_de_clientes['nodo 3'])
    tiempos['nodo 3'].append(tiempo)
    
    
    tiempo = tiempo_actual_simulado
    nuevo_tiempo_simulado = exponencial(valor_lambda)
    
    if(tiempo + nuevo_tiempo_simulado < tiempo_maximo_de_servicio):
        tiempo_mayor['tiempos llegadas'] = tiempo + nuevo_tiempo_simulado
    if(numero_de_clientes['nodo 1'] == 1):
        tiempo_servicio_simulado = normal()
        tiempo_mayor['servicio nodo 1'] = tiempo + tiempo_servicio_simulado
        
```

```python
def servidor_nodo_1(tiempo_actual_simulado):
    global promedio_de_clientes
    global numero_de_clientes
    global tiempo
    global numero_de_salidas
    global salidas
    global numero_de_llegadas
    global llegadas
    global tiempo_mayor
    global clientes, tiempos
    
    
    clientes['nodo 1'].append(numero_de_clientes['nodo 1'])
    tiempos['nodo 1'].append(tiempo)
    clientes['nodo 2'].append(numero_de_clientes['nodo 2'])
    tiempos['nodo 2'].append(tiempo)
    clientes['nodo 3'].append(numero_de_clientes['nodo 3'])
    tiempos['nodo 3'].append(tiempo)
    
    
    promedio_de_clientes['nodo 1'] += numero_de_clientes['nodo 1'] * (tiempo_actual_simulado - tiempo)
    numero_de_clientes['nodo 1'] -= 1
    numero_de_salidas['nodo 1'] += 1
    salidas['nodo 1'].append(tiempo_actual_simulado)
    
    
    clientes['nodo 1'].append(numero_de_clientes['nodo 1'])
    tiempos['nodo 1'].append(tiempo)
    clientes['nodo 2'].append(numero_de_clientes['nodo 2'])
    tiempos['nodo 2'].append(tiempo)
    clientes['nodo 3'].append(numero_de_clientes['nodo 3'])
    tiempos['nodo 3'].append(tiempo)
    
    
    numero_aleatorio = aleatorioRandom()
    if(numero_aleatorio < 0.4):
        promedio_de_clientes['nodo 2'] += numero_de_clientes['nodo 2'] * (tiempo_actual_simulado - tiempo)
        numero_de_clientes['nodo 2'] += 1
        numero_de_llegadas['nodo 2'] += 1
        llegadas['nodo 2'].append(tiempo_actual_simulado)
        if(numero_de_clientes['nodo 2'] == 1):
            nuevo_tiempo_simulado = exponencial(valor_lambda)
            tiempo_mayor['servicio nodo 2'] = tiempo_actual_simulado + nuevo_tiempo_simulado
    else:
        promedio_de_clientes['nodo 3'] += numero_de_clientes['nodo 3'] * (tiempo_actual_simulado - tiempo)
        numero_de_clientes['nodo 3'] += 1
        numero_de_llegadas['nodo 3'] += 1
        llegadas['nodo 3'].append(tiempo_actual_simulado)
        if(numero_de_clientes['nodo 3'] == 1):
            tiempo_servicio_simulado = normal()
            tiempo_mayor['servicio nodo 3'] = tiempo_actual_simulado + tiempo_servicio_simulado
    
    tiempo = tiempo_actual_simulado
    if(numero_de_clientes['nodo 1'] > 0):
        tiempo_servicio_simulado = normal()
        tiempo_mayor['servicio nodo 1'] = tiempo + tiempo_servicio_simulado
        
        
def servidor_nodo_2(tiempo_actual_simulado):
    global promedio_de_clientes
    global numero_de_clientes
    global tiempo
    global numero_de_salidas
    global salidas
    global numero_de_llegadas
    global llegadas
    global tiempo_mayor
    global clientes, tiempos
    
    
    clientes['nodo 1'].append(numero_de_clientes['nodo 1'])
    tiempos['nodo 1'].append(tiempo)
    clientes['nodo 2'].append(numero_de_clientes['nodo 2'])
    tiempos['nodo 2'].append(tiempo)
    clientes['nodo 3'].append(numero_de_clientes['nodo 3'])
    tiempos['nodo 3'].append(tiempo)
    
    
    promedio_de_clientes['nodo 2'] += numero_de_clientes['nodo 2'] * (tiempo_actual_simulado - tiempo)
    numero_de_clientes['nodo 2'] -= 1
    numero_de_salidas['nodo 2'] += 1
    salidas['nodo 2'].append(tiempo_actual_simulado)
    
    
    clientes['nodo 1'].append(numero_de_clientes['nodo 1'])
    tiempos['nodo 1'].append(tiempo)
    clientes['nodo 2'].append(numero_de_clientes['nodo 2'])
    tiempos['nodo 2'].append(tiempo)
    clientes['nodo 3'].append(numero_de_clientes['nodo 3'])
    tiempos['nodo 3'].append(tiempo)
    
    
    if(numero_de_clientes['nodo 2'] > 0):
        nuevo_tiempo_simulado = exponencial(valor_lambda)
        tiempo_mayor['servicio nodo 2'] = tiempo_actual_simulado + nuevo_tiempo_simulado
        
    promedio_de_clientes['nodo 3'] += numero_de_clientes['nodo 3'] * (tiempo_actual_simulado - tiempo)
    numero_de_clientes['nodo 3'] += 1
    numero_de_llegadas['nodo 3'] += 1
    llegadas['nodo 3'].append(tiempo_actual_simulado)
    if(numero_de_clientes['nodo 3'] == 1):
        tiempo_servicio_simulado = normal()
        tiempo_mayor['servicio nodo 3'] = tiempo_actual_simulado + tiempo_servicio_simulado
        

def servidor_nodo_3(tiempo_actual_simulado):
    global promedio_de_clientes
    global numero_de_clientes
    global tiempo
    global numero_de_salidas
    global salidas
    global numero_de_llegadas
    global llegadas
    global tiempo_mayor
    global clientes, tiempos
    
    clientes['nodo 1'].append(numero_de_clientes['nodo 1'])
    tiempos['nodo 1'].append(tiempo)
    clientes['nodo 2'].append(numero_de_clientes['nodo 2'])
    tiempos['nodo 2'].append(tiempo)
    clientes['nodo 3'].append(numero_de_clientes['nodo 3'])
    tiempos['nodo 3'].append(tiempo)
    
    
    promedio_de_clientes['nodo 3'] += numero_de_clientes['nodo 3'] * (tiempo_actual_simulado - tiempo)
    numero_de_clientes['nodo 3'] -= 1
    numero_de_salidas['nodo 3'] += 1
    salidas['nodo 3'].append(tiempo_actual_simulado)
    
    
    clientes['nodo 1'].append(numero_de_clientes['nodo 1'])
    tiempos['nodo 1'].append(tiempo)
    clientes['nodo 2'].append(numero_de_clientes['nodo 2'])
    tiempos['nodo 2'].append(tiempo)
    clientes['nodo 3'].append(numero_de_clientes['nodo 3'])
    tiempos['nodo 3'].append(tiempo)
    
    
    if(numero_de_clientes['nodo 3'] > 0):
        if(numero_de_clientes['nodo 3'] < 5):
            tiempo_servicio_simulado = normal()
        else:
            tiempo_servicio_simulado = normal()
        
        tiempo_mayor['servicio nodo 3'] = tiempo_actual_simulado + nuevo_tiempo_simulado
        
    tiempo = tiempo_actual_simulado


numero_de_simulations = 10
tiempo_simulacion = []
tiempos_despues_cierre = []
tiempo_promedio_sistemas = []

for i in range(numero_de_simulations):
    M = 9999999999 
    tiempo = 0
    tiempo_despues_ultimo_cliente = 0
    actual_simulation_tiempo = 0
    tiempo_maximo_de_servicio = 10
    
    clientes = { # Eje Y
        'nodo 1': [0],
        'nodo 2': [0],
        'nodo 3': [0]
    }
    tiempos = { # Eje X
        'nodo 1': [0],
        'nodo 2': [0],
        'nodo 3': [0]
    }

    tiempo_mayor = {
        'tiempos llegadas': M,
        'servicio nodo 1': M,
        'servicio nodo 2': M,
        'servicio nodo 3': M
    }

    numero_de_llegadas = {
        'nodo 1': 0,
        'nodo 2': 0,
        'nodo 3': 0
    }
    numero_de_salidas = {
        'nodo 1': 0,
        'nodo 2': 0,
        'nodo 3': 0
    }
    numero_de_clientes = {
        'nodo 1': 0,
        'nodo 2': 0,
        'nodo 3': 0
    }

    llegadas = {
        'nodo 1': [],
        'nodo 2': [],
        'nodo 3': []
    }
    salidas = {
        'nodo 1': [],
        'nodo 2': [],
        'nodo 3': []
    }

    promedio_de_clientes = {
        'nodo 1': 0,
        'nodo 2': 0,
        'nodo 3': 0
    }

    tiempo_promedio_sistema = 0
    valor_lambda = 2


    tiempo_actual_simulado = exponencial(valor_lambda)

    if(tiempo_actual_simulado > tiempo_maximo_de_servicio):
        tiempo_despues_ultimo_cliente = 0
        tiempo_promedio_sistema = 0
        promedio_de_clientes['nodo 1'] = 0
        promedio_de_clientes['nodo 2'] = 0
        promedio_de_clientes['nodo 3'] = 0
    else:
        llegada(tiempo_actual_simulado)
        contador = 0
        while(tiempo_mayor['tiempos llegadas'] != M
              or tiempo_mayor['servicio nodo 1'] != M
              or tiempo_mayor['servicio nodo 2'] != M
              or tiempo_mayor['servicio nodo 3'] != M):
                
            if(min(tiempo_mayor, key=lambda x: tiempo_mayor[x]) == 'tiempos llegadas'):
                nuevo_tiempo_simulado = tiempo_mayor['tiempos llegadas']
                tiempo_mayor['tiempos llegadas'] = M
                llegada(nuevo_tiempo_simulado)
            if(min(tiempo_mayor, key=lambda x: tiempo_mayor[x]) == 'servicio nodo 1'):
                nuevo_tiempo_simulado = tiempo_mayor['servicio nodo 1']
                tiempo_mayor['servicio nodo 1'] = M
                servidor_nodo_1(nuevo_tiempo_simulado)
            if(min(tiempo_mayor, key=lambda x: tiempo_mayor[x]) == 'servicio nodo 2'):
                nuevo_tiempo_simulado = tiempo_mayor['servicio nodo 2']
                tiempo_mayor['servicio nodo 2'] = M
                servidor_nodo_2(nuevo_tiempo_simulado)
            if(min(tiempo_mayor, key=lambda x: tiempo_mayor[x]) == 'servicio nodo 3'):
                nuevo_tiempo_simulado = tiempo_mayor['servicio nodo 3']
                tiempo_mayor['servicio nodo 3'] = M
                servidor_nodo_3(nuevo_tiempo_simulado)

        tiempo_despues_ultimo_cliente = max(0, tiempo - tiempo_maximo_de_servicio)
        acum_1 = 0
        acum_2 = 0
        acum_3 = 0

        contador = 0
        while(contador < numero_de_llegadas['nodo 1']):
            acum_1 += salidas['nodo 1'][contador] - llegadas['nodo 1'][contador]
            contador += 1

        contador = 0
        while(contador < numero_de_llegadas['nodo 2']):
            acum_2 += salidas['nodo 2'][contador] - llegadas['nodo 2'][contador]
            contador += 1

        contador = 0
        while(contador < numero_de_llegadas['nodo 3']):
            acum_3 += salidas['nodo 3'][contador] - llegadas['nodo 3'][contador]
            contador += 1

        try:
            tiempo_promedio_sistema = ((acum_1  / numero_de_llegadas['nodo 1'])
                    + (0.4 * acum_2  / numero_de_llegadas['nodo 2']) 
                    + (acum_3  / numero_de_llegadas['nodo 3']))
        except Exception:
            pass
        

        promedio_de_clientes['nodo 1'] /= tiempo
        promedio_de_clientes['nodo 2'] /= tiempo
        promedio_de_clientes['nodo 3'] /= tiempo
        
        tiempo_promedio_sistemas.append(tiempo_promedio_sistema)
        tiempo_simulacion.append(tiempo)
        tiempos_despues_cierre.append(tiempo_despues_ultimo_cliente)

totalClientes=len(llegadas['nodo 1'])+len(llegadas['nodo 2'])+len(llegadas['nodo 3'])

print("Clientes atendidos: ",(totalClientes))
print("Numero clientes '<<Nodo 1>>: {}".format(numero_de_llegadas['nodo 1']))
print("Numero clientes '<<Nodo 2>>: {}".format(numero_de_llegadas['nodo 2']))
print("Numero clientes '<<Nodo 3>>: {}".format(numero_de_llegadas['nodo 3']))
print("Simulaciones realizadas: {}".format(len(tiempo_simulacion)))
print("Tiempo de simulación promedio: {}".format(sum(tiempo_simulacion) / len(tiempo_simulacion)))
print("Tiempo medio del sistema: {}".format(sum(tiempo_promedio_sistemas) / len(tiempo_promedio_sistemas)))
```
